## Combine each country dataframe into one dataframe

```python
country1=pd.concat([new_CA_df,new_DE_df])
country2=pd.concat([country1,new_FR_df])
country3=pd.concat([country2,new_GB_df])
youtube_trending_df=pd.concat([country3,new_US_df])
youtube_trending_df.head()
```

```python
csv_file = "sources/category_id.csv"
category_id_df = pd.read_csv(csv_file)
category_id_df.head()
```

```python
category_id_df.rename(columns={"id":"category_id"},inplace=True)
category_id_df.head()
```

```python
category_id_df.dtypes
```

```python
population_csv_file = "sources/world_population.csv"
world_population_df = pd.read_csv(population_csv_file)
world_population_df.head()
```

```python
population_df = world_population_df[world_population_df['Country'].isin(['Canada','United States of America','United Kingdom','France','Germany'])]
population_df
```

```python
new_population_df= population_df[['Country','Population']]
new_population_df
```

```python
#new_population_df=new_population_df.set_index("Country")
```

```python
country_id_list=[0,1,2,3,4]
new_population_df["country_id"]=country_id_list
#new_population_df=new_population_df.set_index("country_id")
new_population_df
```

```python
new_population_df.rename(columns={"Country":"country","Population":"population"},inplace=True)
new_population_df
```

```python
merged_df= pd.merge(youtube_trending_df,new_population_df, on =['country_id'])
new_youtube_df= pd.merge(merged_df,category_id_df, on =['category_id'])
```

```python
new_youtube_df.head()
```

## Connect to local database

```python
rds_connection_string = "root:root@127.0.0.1/youtube_db"
engine = create_engine(f'mysql://{rds_connection_string}')
```

## Check for tables

```python
engine.table_names()
```

```python
youtube_trending_df.to_sql(name='youtube', con=engine, if_exists='append', index=False)
```

```python
new_population_df.to_sql(name='population', con=engine, if_exists='append', index=False)
```

```python
category_id_df.to_sql(name='category', con=engine, if_exists='append', index=False)
```

```python
new_youtube_df.to_sql(name='youtube_trending', con=engine, if_exists='append', index=False)
```
